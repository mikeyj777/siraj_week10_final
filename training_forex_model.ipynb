```python
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression
from sklearn.kernel_ridge import KernelRidge
from sklearn.svm import SVR
from sklearn.externals import joblib
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df = pd.read_csv('data/fxdata1.csv')
```

```python
df.head()
```

```python
dates = np.asarray(df['Date'], dtype = np.float64)

print(dates)

currenciesallrows = np.asarray(df['Currency'])

print(currenciesallrows)

currencies = np.unique(currenciesallrows)

print(currencies)

costsOf1USD = np.asarray(df['CostOf1USD'], dtype = np.float64)

print(costsOf1USD)
```

```python
paramsdict = {}
scores = []
for currency in currencies:
    print(currency)
    X = dates[currenciesallrows == currency].reshape(-1, 1)
    y = costsOf1USD[currenciesallrows == currency]
    clf = SVR()
    print(clf, ' instantiated')
#     X,y = shuffle(X,y)
#     print('shuffled')
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
    print('split')
    clf.fit(X_train,y_train)
    print('trained')
    a = clf.score(X_test,y_test)
    plt.plot(X_test,y_test,color = 'red', alpha = 0.5)
    plt.plot(X_test,clf.predict(X_test),color = 'blue', alpha = 0.7)
    plt.show()
    print(currency, clf.score(X_test,y_test))
    scores.append((currency, clf.score(X_test,y_test)))
#     paramsdict[currency] = (clf.coef_, clf.intercept_)
    filenm = 'data/fxpredsklearn_SVR_USD_' + currency + '.sav'
    joblib.dump(clf,filenm)

```

```python
scores = np.asarray(scores)
print(scores.shape)
print(scores)
```
